```python
import os
from dotenv import load_dotenv
import kagglehub

# Load environment variables from .env file
load_dotenv()

# kagglehub automatically picks up KAGGLE_API_TOKEN from the environment variables
# Download latest version of the dataset
path = kagglehub.dataset_download("hasindukoshitha/sinhala-ocr-training-data")
```

```python
import shutil
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

data_dir = os.getenv("DATA_DIR", "./datasets/train_images")
csv_path = os.getenv("CSV_PATH", "./datasets/metadata.csv")

# Create parent directories if they don't exist
os.makedirs(os.path.dirname(data_dir), exist_ok=True)
os.makedirs(os.path.dirname(csv_path), exist_ok=True)

try:
    # Move train_images folder
    train_images_src = os.path.join(path, "train_images")
    if os.path.exists(train_images_src):
        if os.path.exists(data_dir):
            shutil.rmtree(data_dir)
        shutil.move(train_images_src, data_dir)
        print(f"✓ train_images moved to: {data_dir}")
    else:
        print(f"✗ train_images not found in {path}")
    
    # Move metadata.csv file
    csv_src = os.path.join(path, "metadata.csv")
    if os.path.exists(csv_src):
        if os.path.exists(csv_path):
            os.remove(csv_path)
        shutil.move(csv_src, csv_path)
        print(f"✓ metadata.csv moved to: {csv_path}")
    else:
        print(f"✗ metadata.csv not found in {path}")
    
    # Verify
    if os.path.exists(data_dir):
        print(f"✓ Verification: {data_dir} exists with {len(os.listdir(data_dir))} items")
    if os.path.exists(csv_path):
        print(f"✓ Verification: {csv_path} exists")
        
except Exception as e:
    print(f"✗ ERROR: {e}")
```

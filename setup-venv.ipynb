### 1. Create the Virtual Environment
This creates a folder named `venv` in your directory.

```python
!python -m venv venv
```

### 2. Install requirements using the venv's pip
Running `!activate` in a notebook cell **does not work** because `!` runs in a temporary sub-shell that immediately closes. Instead, we call the `pip.exe` directly from the `venv` we just created.

```python
!.\venv\Scripts\pip.exe install -r requirements.txt
```

### 4. Register the venv as a Jupyter Kernel
To actually use this virtual environment in Jupyter notebooks, we need to add it as a kernel. After running this, **refresh your kernel list in VSCode/Jupyter** (top right corner) and select `venv_trocr`.

```python
!.\venv\Scripts\pip.exe install ipykernel
!.\venv\Scripts\python.exe -m ipykernel install --user --name=venv_trocr
```
